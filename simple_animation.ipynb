# Simple Animation
This notebook simulates particles wobbling around randomly, i.e. the most simple version of motion.

Contact: simtho@kth.se

## Import Modules

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import matplotlib.animation as animation
%matplotlib notebook
```

## Variable Declaration

```python
SIZE = (3, 100)
save_video = False
```

# Computation and Visualization

```python
a = np.array(np.random.uniform(-10, 10, size=SIZE))

scatters = []
fig = plt.figure()
ax = plt.axes(projection='3d')
 
for i in range(100):
    a += np.random.normal(0, 0.1, size=SIZE)
    scatter = ax.scatter(a[0], a[1], a[2], c="b", animated=True)
    scatters.append([scatter])

#ax.set_xlim(0, 10)
#ax.set_ylim(0, 10)
#ax.set_zlim(0, 10)
ani = animation.ArtistAnimation(fig, scatters, interval=50, blit=True, repeat=False)

if save_video:
    ani.save('results/moving_particles.mp4', dpi=300)
```
